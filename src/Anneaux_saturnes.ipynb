```python
#constante Gravitationnelle (m**3*kg**(-1)*s**2)
G= 6.6742*10**(-11)
#Masse de Saturne
Ms=5.683*10**26
#rayon de Saturne (km)
r_saturne=58232
#rayon min (km)
r_min=60000
#rayon max (km)
r_max=302500
B_o=1
#masse d'une particule (kg)
mp=1
#charge (C)
q=1
#vitesse initiale (m/s)
v_int=0.1
#angle particule (radian)
a_p=0.5
#position x initiale
x_int=-0.1
#position y initiale
y_int=0
#position z initiale
z_int=0
#temps maximal
t_max=100

w=(q*B_o)/mp
```

```python
def rayon(r_min, r_max):
    """int*int->int"""
    return np.random.randint(r_min,r_max)*10**3
```

```python
r=rayon (r_min, r_max)
```

```python
def acceleration(G,Ms,r):
    """float*float*int*int->float"""
    return (-G*Ms)/r**2
```

```python
acceleration(G,Ms,r)
```

```python
def vitesse(G,Ms,r):
    """float*float*int*int->float"""
    return (np.sqrt((G*Ms)/(r)))
    
```

```python
vitesse(G,Ms,r)
```

```python
def periode(G,Ms,r):
    """float*float*int*int->float"""
    #vitesse:float
    v=vitesse(G,Ms,r)
    return (np.pi*2)*r/(v*86400)
```

```python
p=periode(G,Ms,r)*86400
```

```python
def angle(G,Ms,r,t):
    #periode
    p=periode(G,Ms,r)
    #vitesse angulaire
    omg=(np.pi*2)/(p*86400)
    return (omg*t)*(180/np.pi)
```

```python
angle (G,Ms,r,p)
```

```python
def list_t(G,Ms,r):
    a=angle(G,Ms,r,0)
    i=60
    theta=[]
    while (a<=360.0):
        theta.append(a)
        i=i+60
        a=angle(G,Ms,r,i)
        
    return theta

theta = list_t(G,Ms,r)
x1 = r*np.cos(theta)
x2 = r*np.sin(theta)

fig, ax = plt.subplots(1)
plt.title("trajectoire des particules")
ax.plot(x1, x2)
ax.set_aspect(1)
plt.grid(linestyle='--')
```

```python
def miroir_x(B_o,mp,q,v_int,a_p,x_int,w,t_max):
    t=0
    x=0
    liste_x=[]
    while (t<=100):
        x=x_int +(v_int*math.cos(a_p)*math.sin(w*t))/w
        t=t+5
        liste_x.append(x)
    return liste_x
def miroir_y(B_o,mp,q,v_int,a_p,y_int,w,t_max):
    t=0
    y=0
    liste_y=[]
    while (t<=100):
        y=y_int -(v_int*math.cos(a_p)*math.cos(w*t))/w
        t=t+5
        liste_y.append(y)
    return liste_y

def miroir_z(B_o,mp,q,v_int,a_p,z_int,w,t_max):
    t=0
    z=0
    liste_z=[]
    while (t<=100):
        z=v_int*math.sin(a_p)*t
        t=t+5
        liste_z.append(z)
    return liste_z
```

```python
liste_x =miroir_x(B_o,mp,q,v_int,a_p,x_int,w,t_max)
liste_y =miroir_y(B_o,mp,q,v_int,a_p,y_int,w,t_max)
liste_z =miroir_z(B_o,mp,q,v_int,a_p,z_int,w,t_max)
```

```python
fig, ax = plt.subplots(1)
ax.plot(liste_x, liste_y,liste_z)
ax.set_aspect(1)
plt.grid(linestyle='--')

ax=Axes3D(fig)
plt.gca(projection='3d')
plt.plot(liste_x, liste_y, liste_z)
plt.show()
```
